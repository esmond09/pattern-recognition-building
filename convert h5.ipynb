```python
import h5py
import numpy as np
import os
from PIL import Image
```

```python
import cv2
def remove_noise(image):
    denoised_image = cv2.fastNlMeansDenoising(np.array(image), None, h=10, templateWindowSize=7, searchWindowSize=21)
    return Image.fromarray(denoised_image)
```

```python
subfile1_path = "/content/drive/MyDrive/pattern recognition photos/pr dataset1/afamosa"
subfile2_path = "/content/drive/MyDrive/pattern recognition photos/pr dataset1/chris church"
subfile3_path = "/content/drive/MyDrive/pattern recognition photos/pr dataset1/saint paul church"
output_h5_file = "/content/drive/MyDrive/pattern recognition photos/pr_dataset.h5"
```

```python
images = []
labels = []
```

```python
subfile1_folder = os.path.join(subfile1_path)
class1_label = 0  # Assign a label to the images in the first sub-file

for image_file in os.listdir(subfile1_folder):
    image_path = os.path.join(subfile1_folder, image_file)
    image = Image.open(image_path)
    image = image.convert("L")
    image = image.resize((256, 256))
    image = remove_noise(image)
    image_array = np.array(image)
    images.append(image_array)
    labels.append(class1_label)
```

```python
subfile2_folder = os.path.join(subfile2_path)
class2_label = 1  # Assign a label to the images in the second sub-file

for image_file in os.listdir(subfile2_folder):
    image_path = os.path.join(subfile2_folder, image_file)
    image = Image.open(image_path)
    image = image.convert("L")
    image = image.resize((256, 256))
    image = remove_noise(image)
    image_array = np.array(image)
    images.append(image_array)
    labels.append(class2_label)
```

```python
subfile3_folder = os.path.join(subfile3_path)
class3_label = 2  # Assign a label to the images in the third sub-file

for image_file in os.listdir(subfile3_folder):
    image_path = os.path.join(subfile3_folder, image_file)
    image = Image.open(image_path)
    image = image.convert("L")
    image = image.resize((256, 256))
    image = remove_noise(image)
    image_array = np.array(image)
    images.append(image_array)
    labels.append(class3_label)
```

```python
images = np.array(images)
labels = np.array(labels)
```

```python
with h5py.File(output_h5_file, "w") as hf:
    hf.create_dataset("images", data=images)
    hf.create_dataset("labels", data=labels)
```
